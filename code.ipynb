```python
import matplotlib.pyplot as plt
categories = ['math.grade', 'english.grade', 'sciences.grade', 'language.grade', 'cover letter','refer letter']
values = [20, 15, 30, 35, 10, 18]


plt.figure(figsize=(10, 6))
plt.barh(categories, values, color=['lightblue', 'pink'])
plt.xlabel("subject")
plt.ylabel("No. of subject")
plt.title("Report")
plt.grid()
plt.show()
```
